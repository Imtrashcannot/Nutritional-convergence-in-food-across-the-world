```python
import pandas as pd

file_path = "en.openfoodfacts.org.products.csv"

# Load only the first 100 rows to see the structure without consuming RAM
df_sample = pd.read_csv(file_path, sep='\t', nrows=100)

print("--- Columns List ---")
print(df_sample.columns.tolist())
display(df_sample.head())
display(df_sample.tail())

display(df_sample.describe())
```

```python
import numpy as np

# 1. We only need these cols
essential_cols = [
    'product_name', 'generic_name', 'countries_en', 'categories_en', 
    'nutriscore_grade', 'nova_group',
    'energy-kcal_100g', 'fat_100g', 'saturated-fat_100g', 
    'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 
    'salt_100g', 'fiber_100g'
]

# 2. Targeting SEA and EA vs EU and Europe
asia_list = ['China', 'Japan', 'South Korea', 'Mongolia', 'Taiwan', 'Brunei', 'Cambodia', 'Indonesia', 
             'Laos', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam']
eu_list = ['Austria', 'Belgium', 'Denmark', 'Finland', 'France', 'Germany', 'Ireland', 'Luxembourg', 'Netherlands', 'Sweden', 
           'Greece', 'Cyprus', 'Italy', 'Malta', 'Portugal', 'Spain', 'Bulgaria', 'Croatia', 'Czech Republic', 'Estonia', 'Hungrary', 
           'Latvia', 'Lithuania', 'Poland', 'Romania', 'Slovakia', 'Slovenia', 'United Kingdom', 'Ukraine', 'Russia']

# 3. Processing, the potato laptop should be able to survive this... right?
file_path = "en.openfoodfacts.org.products.csv"
chunk_size = 30000
output_list = []

print("Starting efficient extraction...")

# ChatGPT did this for me
for chunk in pd.read_csv(file_path, sep='\t', usecols=essential_cols, chunksize=chunk_size, low_memory=False):
    
    # Filter for Asia or Europe immediately
    # We use .str.contains to catch "Japan, France" etc.
    mask_asia = chunk['countries_en'].str.contains('|'.join(asia_list), na=False, case=False)
    mask_eu = chunk['countries_en'].str.contains('|'.join(eu_list), na=False, case=False)
    
    # Create the 'Region' tag
    chunk_filtered = chunk[mask_asia | mask_eu].copy()
    chunk_filtered['Region'] = np.where(chunk_filtered['countries_en'].str.contains('|'.join(asia_list), case=False), 'Asia', 'Europe')
    
    # Drop rows where all nutrition data is missing
    chunk_filtered = chunk_filtered.dropna(subset=['energy-kcal_100g', 'proteins_100g'], how='all')
    
    output_list.append(chunk_filtered)

# 4. Saveeee
df_final = pd.concat(output_list)
df_final.to_csv("asia_eu_food_comparison.csv", index=False)

print(f"Done! Created a small file with {len(df_final)} rows.")
print(df_final['Region'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("asia_eu_food_comparison.csv", encoding='utf-8')

# Drop rows with NaN in core features for PCA
features = ['energy-kcal_100g', 'fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 'salt_100g']
for c in features:
    df[c] = (df[c].astype(str)
                  .str.replace(",", ".", regex=False)
                  .replace({"nan": np.nan, "None": np.nan, "": np.nan}))
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Now, let us remove some outliers
# 1. Energy cannot exceed 900 kcal per 100g (Pure Fat is 900-ish)
df = df[df['energy-kcal_100g'] <= 900]

# 2. Sum of macros cannot exceed 100g
macro_sum = df['fat_100g'].fillna(0) + df['carbohydrates_100g'].fillna(0) + df['proteins_100g'].fillna(0)
df = df[macro_sum <= 100]

# 3. Individual components cannot be negative or > 100
for col in ['fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 'salt_100g']:
    df = df[(df[col] >= 0) & (df[col] <= 100)]

# 4. Salt check (anything > 50g salt is likely a mistake or pure salt)
df = df[df['salt_100g'] <= 50]

# cleaning
df_clean = df.dropna(subset=features).copy()
print(f"Data remaining after cleaning: {len(df_clean)}")
```

```python
# now we scale and PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_clean[features])

pca = PCA(n_components=2)
pca_results = pca.fit_transform(scaled_data)

df_clean['pca_1'] = pca_results[:, 0]
df_clean['pca_2'] = pca_results[:, 1]

# this comment onward is for visualization
# Sample 10k from each to see the overlap/difference clearly
asia_sample = df_clean[df_clean['Region'] == 'Asia'].sample(min(10000, len(df_clean[df_clean['Region'] == 'Asia'])))
eu_sample = df_clean[df_clean['Region'] == 'Europe'].sample(min(10000, len(df_clean[df_clean['Region'] == 'Europe'])))
plot_df = pd.concat([asia_sample, eu_sample])

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=plot_df, x='pca_1', y='pca_2', 
    hue='Region', alpha=0.4, palette={'Asia': '#FF4B4B', 'Europe': '#4B4BFF'}
)

# Limit axes to remove any remaining extreme outliers from the plot view
plt.xlim(df_clean['pca_1'].quantile(0.01), df_clean['pca_1'].quantile(0.99))
plt.ylim(df_clean['pca_2'].quantile(0.01), df_clean['pca_2'].quantile(0.99))

plt.title("Cleaned Food Galaxy: Asia vs Europe")

explained_var = pca.explained_variance_ratio_
total_var = explained_var.sum()
print(f"PCA 1 explains: {explained_var[0]:.2%}")
print(f"PCA 2 explains: {explained_var[1]:.2%}")
print(f"Total variance captured by 2D: {total_var:.2%}")

plt.show()

loadings = pd.DataFrame(
    pca.components_.T, 
    columns=['PCA1', 'PCA2'], 
    index=features
)
print("--- PCA Loadings (How much each nutrient affects the axis) ---")
print(loadings)
```

```python
plt.figure(figsize=(12, 8))

# Plotting the "Density" of European food (Blue)
sns.kdeplot(
    data=df_clean[df_clean['Region']=='Europe'].sample(20000), 
    x='pca_1', y='pca_2', 
    cmap="Blues", fill=True, alpha=0.5, thresh=0.1, label='Europe Density'
)

# Plotting the "Density" of Asian food (Red)
sns.kdeplot(
    data=df_clean[df_clean['Region']=='Asia'].sample(20000), 
    x='pca_1', y='pca_2', 
    cmap="Reds", fill=True, alpha=0.5, thresh=0.1, label='Asia Density'
)

plt.title("Nutritional Fingerprints: Europe vs Asia")
plt.xlabel(f"PCA1 ({explained_var[0]:.1%})")
plt.ylabel(f"PCA2 ({explained_var[1]:.1%})")
plt.legend()
plt.show()
```

```python
from scipy import stats

results = []
for nutrient in ['energy-kcal_100g', 'fat_100g', 'sugars_100g', 'salt_100g']:
    asia_vals = df_clean[df_clean['Region'] == 'Asia'][nutrient]
    eu_vals = df_clean[df_clean['Region'] == 'Europe'][nutrient]

    # Kruskal Wallis test
    stat, p_value = stats.kruskal(asia_vals, eu_vals)

    # effect size (Cliff's Delta)
    mean_diff = (asia_vals.mean() - eu_vals.mean())

    results.append({
        'Nutrient': nutrient,
        'P-value': p_value,
        'Mean Difference': round(mean_diff, 2),
        'Interpretation': "Statistically Different" if p_value < 0.05 else "No Different"
    })

stats_df = pd.DataFrame(results)
print(stats_df)
```

```python
# 1. Cleaning the NOVA column (it's often 1.0, 2.0, etc. or NaN)
df_clean['nova_group'] = pd.to_numeric(df_clean['nova_group'], errors='coerce')
df_nova = df_clean.dropna(subset=['nova_group'])

# 2. Plotting the PCA map but colored by Processed Level
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df_nova.sample(20000), x='pca_1', y='pca_2', 
    hue='nova_group', palette='viridis', alpha=0.5
)
plt.title("The Industrialization of Food: PCA by NOVA Group")
plt.show()

# 3. Comparing Asia vs Europe by Processed Level Distribution
nova_comparison = pd.crosstab(df_nova['Region'], df_nova['nova_group'], normalize='index') * 100
print("--- Percentage of Products in each Processing Group (%) ---")
display(nova_comparison.round(2))
```

```python
# 1. Filter for minimally processed and culinary ingredients only
df_basics = df_clean[df_clean['nova_group'].isin([1.0, 2.0])].copy()

print(f"Items remaining for 'Basics' analysis: {len(df_basics)}")
print(f"Distribution: \n{df_basics['Region'].value_counts()}")

# 2. Re-run the Statistical Test for this subset
results_basics = []
for nutrient in ['energy-kcal_100g', 'sugars_100g', 'fat_100g', 'salt_100g']:
    asia_vals = df_basics[df_basics['Region'] == 'Asia'][nutrient]
    eu_vals = df_basics[df_basics['Region'] == 'Europe'][nutrient]
    
    stat, p_value = stats.kruskal(asia_vals, eu_vals)
    mean_diff = asia_vals.mean() - eu_vals.mean()
    
    results_basics.append({
        'Nutrient': nutrient,
        'Mean Diff (Asia-EU)': round(mean_diff, 2),
        'P-Value': p_value
    })

print("\n--- Statistical Comparison of NOVA 1 & 2 ---")
display(pd.DataFrame(results_basics))
```

```python
# 1. Filter for Basics
df_basics = df_clean[df_clean['nova_group'].isin([1.0, 2.0])].copy()

# 2. IQR Cleaning (Removes the "Typo" Outliers)
# We do this per nutrient to be safe
for nutrient in features:
    Q1 = df_basics[nutrient].quantile(0.25)
    Q3 = df_basics[nutrient].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # We only filter the upper bound usually, as nutrients can be 0 
    # but shouldn't be "impossible" for the category
    df_basics = df_basics[(df_basics[nutrient] >= 0) & (df_basics[nutrient] <= upper_bound)]

print(f"Basics after Statistical Cleaning: {len(df_basics)}")
```

```python
# 1. Standardize and PCA
scaler = StandardScaler()
scaled_basics = scaler.fit_transform(df_basics[features])

pca_basics = PCA(n_components=2)
pca_results = pca_basics.fit_transform(scaled_basics)

df_basics['pca_1'] = pca_results[:, 0]
df_basics['pca_2'] = pca_results[:, 1]

# 2. Calculate Variance & Loadings
exp_var = pca_basics.explained_variance_ratio_
loadings = pd.DataFrame(pca_basics.components_.T, columns=['PCA1', 'PCA2'], index=features)

print(f"PCA Energy: {exp_var.sum():.2%}")
print("\n--- Loadings (What drives the 'Basics' map) ---")
print(loadings)

# 3. Plotting the Cleaned Basics
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df_basics, x='pca_1', y='pca_2', 
    hue='Region', palette={'Asia': 'red', 'Europe': 'blue'}, alpha=0.5
)
plt.title("NOVA 1 & 2: Regional Nutritional Foundations (Cleaned)")
plt.show()
```

```python
# 1. Calculate the Means for each region (NOVA 1 & 2 only)
# Features: energy, fat, carbs, sugars, proteins, salt
features = ['energy-kcal_100g', 'fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 'salt_100g']
# features = ['fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 'salt_100g']

asia_means = df_basics[df_basics['Region'] == 'Asia'][features].mean()
eu_means = df_basics[df_basics['Region'] == 'Europe'][features].mean()

# 2. Prepare the data for plotting
plot_df = pd.DataFrame({
    'Nutrient': features,
    'Asia (Traditional)': asia_means.values,
    'Europe (Traditional)': eu_means.values
}).melt(id_vars='Nutrient', var_name='Region', value_name='Average Value (per 100g)')

# 3. Create the Visualization
plt.figure(figsize=(12, 7))
sns.barplot(data=plot_df, x='Nutrient', y='Average Value (per 100g)', hue='Region', palette=['#FF4B4B', '#4B4BFF'])

# Add labels and formatting
plt.title("Regional Nutritional Profiles: Comparing Basic Ingredients (NOVA 1 & 2)", fontsize=14)
plt.ylabel("Mean Value (kcal or grams)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars for precision
for i, p in enumerate(plt.gca().patches):
    plt.gca().annotate(f'{p.get_height():.1f}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                   fontsize=10)

plt.tight_layout()


plt.savefig("regional nutritional profiles.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
results_final = []
for nutrient in features:
    asia_vals = df_basics[df_basics['Region'] == 'Asia'][nutrient]
    eu_vals = df_basics[df_basics['Region'] == 'Europe'][nutrient]
    
    # Kruskal-Wallis
    stat, p_value = stats.kruskal(asia_vals, eu_vals)
    
    results_final.append({
        'Nutrient': nutrient,
        'Asia Mean': round(asia_vals.mean(), 2),
        'EU Mean': round(eu_vals.mean(), 2),
        'P-Value': p_value,
        'Significant?': "Yes" if p_value < 0.05 else "No"
    })

final_stats_df = pd.DataFrame(results_final)
print("--- Final Statistical Proof (NOVA 1 & 2) ---")
display(final_stats_df)
```
